```python
import os
```

```python
def mkdir(p):
    if not os.path.exists(p):
        os.mkdir(p)
        
def link(src, dst):
    if not os.path.exists(dst):
        os.symlink(src, dst, target_is_directory=True)
```

- now we build out the VGG network

```python
from __future__ import print_function, division
from builtins import range, input

from keras.layers import Input, Lambda, Dense, Flatten
from keras.models import Model

#### algorithms

# vgg
#from keras.applications.vgg16 import VGG16, preprocess_input

# resnet
from keras.applications.resnet50 import ResNet50, preprocess_input

#### algorithms

from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

from glob import glob

from PIL import Image

%matplotlib inline
```

```python
# vgg
#IMAGE_SIZE = [100, 100]

# resnet (note: minimum resnet image size is 197x197)
IMAGE_SIZE = [224,224]
epochs = 1
batch_size = 32

train_path = './data/large_files/fruits-360-small/Training'
valid_path = './data/large_files/fruits-360-small/Validation'

image_files = glob(train_path + '/*/*.jp*g')
valid_image_files = glob(valid_path + '/*/*.jp*g')

folders = glob(train_path + '/*')
folders
```

```python

plt.imshow(image.load_img(np.random.choice(image_files)))
plt.show()
```

```python
# preprocessing for front of VGG
# remember VGG is a cone shaped repetive list of convolutions so
# the include_top=false will leave off that top which we're going to
# customize

#vgg algo
#algo = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)

# resnet algo
algo = ResNet50(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)

# turn off training for all but that top layer
for layer in algo.layers:
    layer.trainable = False
    
x = Flatten()(algo.output)

prediction = Dense(len(folders), activation='softmax')(x)

# prediction is the top layer (training is turned on for it)
model = Model(inputs=algo.input, outputs=prediction)

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
gen = ImageDataGenerator(rotation_range=20, width_shift_range=0.1,
                        height_shift_range=0.1, shear_range=0.1, zoom_range=0.2,
                        horizontal_flip=True, vertical_flip=True,
                        preprocessing_function=preprocess_input)
```

```python
# get label mapping for confusion matrix
test_gen = gen.flow_from_directory(valid_path, target_size=IMAGE_SIZE)
print(test_gen.class_indices)
labels = [None] * len(test_gen.class_indices)

for k, v in test_gen.class_indices.items():
    labels[v] = k
```

```python
# image should have strange colors due to VGG weights being BGR
for x,y in test_gen:
    print("min:", x[0].min(), "max:", x[0].max())
    plt.title(labels[np.argmax(y[0])])
    plt.imshow(x[0])
    plt.show()
    break
```

```python
# create generators

train_generator = gen.flow_from_directory(train_path, target_size=IMAGE_SIZE,
                                         shuffle=True, batch_size=batch_size)
valid_generator = gen.flow_from_directory(valid_path, target_size=IMAGE_SIZE,
                                         shuffle=True, batch_size=batch_size)
```

```python
# fit the model

r = model.fit_generator(train_generator, validation_data=valid_generator,
                       epochs=epochs, steps_per_epoch=len(image_files) // batch_size,
                       validation_steps=len(valid_image_files) // batch_size)
```

```python
def get_confusion_matrix(data_path, N):
    print("Generating confusion matrix", N)
    predictions = []
    targets = []
    i = 0
    for x, y in gen.flow_from_directory(data_path, target_size=IMAGE_SIZE,
                                       shuffle=False, batch_size=batch_size * 2):
        i += 1
        if i % 50 == 0:
            print(i)
        p = model.predict(x)
        p = np.argmax(p, axis=1)
        y = np.argmax(y, axis=1)
        predictions = np.concatenate((predictions, p))
        targets = np.concatenate((targets, y))
        if len(targets) >= N:
            # it will keep looping otherwise
            break
            
    cm = confusion_matrix(targets, predictions)
    return cm
```

```python
cm = get_confusion_matrix(train_path, len(image_files))
print(cm)
valid_cm = get_confusion_matrix(valid_path, len(valid_image_files))
print(valid_cm)
```

```python
# loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()

# accuracies
plt.plot(r.history['acc'], label='train acc')
plt.plot(r.history['val_acc'], label='val acc')
plt.legend()
plt.show()

from util import plot_confusion_matrix
plot_confusion_matrix(cm, labels, title='Train confusion matrix')
plot_confusion_matrix(valid_cm, labels, title='Validation confusion matrix')
```
